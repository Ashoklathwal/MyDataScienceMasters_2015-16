#Assignment, question 2 - Creating category-specific image retrieval models

```python
# Split the image_train data SFrame into 4 different SFrames, one for each categorie  {‘dog’,’cat’,’automobile’,bird’}
dog = image_train[image_train['label'] == 'dog']
cat = image_train[image_train['label'] == 'cat']
automobile = image_train[image_train['label'] == 'automobile']
bird = image_train[image_train['label'] == 'bird']
```

```python
bird['image'].show() # Test
```

```python
#### create a nearest neighbor model using the 'deep_features' as the features, one model for each category
dog_model = graphlab.nearest_neighbors.create(dog, features = ['deep_features'],
                                             label = 'id')
cat_model = graphlab.nearest_neighbors.create(cat, features = ['deep_features'],
                                             label = 'id')
automobile_model = graphlab.nearest_neighbors.create(automobile, features = ['deep_features'],
                                             label = 'id')
bird_model = graphlab.nearest_neighbors.create(bird, features = ['deep_features'],
                                             label = 'id')
```

```python
# Nearest ‘cat’ labeled image in the training data to the first cat image in the test data
image_test = graphlab.SFrame('image_test_data')
cat_model.query(image_test[0:1])
```

#Assignment, question 4 -  Computing nearest neighbors accuracy using SFrame operations

```python
# Spliting test data by label
image_test_cat = image_test[image_test['label'] == 'cat']
image_test_dog = image_test[image_test['label'] == 'dog']
image_test_bird = image_test[image_test['label'] == 'bird']
image_test_automobile = image_test[image_test['label'] == 'automobile']
```

```python
# find the closest neighbor in to the dog test data using each of the trained models
dog_cat_neighbors = cat_model.query(image_test_dog, k=1)
dog_dog_neighbors = dog_model.query(image_test_dog, k=1)
dog_bird_neighbors = bird_model.query(image_test_dog, k=1)
dog_automobile_neighbors = automobile_model.query(image_test_dog, k=1)
```

```python
dog_cat_neighbors.head()
```

```python
# Create a new SFrame with the distances from ‘dog’ test examples to the respective nearest neighbors in each class in the training data
dog_distances = graphlab.SFrame({'dog-automobile': dog_automobile_neighbors['distance'], 'dog-bird': dog_bird_neighbors['distance'], 'dog-cat': 
                                 dog_cat_neighbors['distance'], 'dog-dog': dog_dog_neighbors['distance']})
```

```python
dog_distances.head()
```

```python
# Create a function that returns 1 if the value for row[‘dog-dog’] is lower than that of the other columns, and 0 otherwise
def is_dog_correct(row):
    if row['dog-dog'] < row['dog-cat'] and row['dog-dog'] < row['dog-bird'] and row['dog-dog'] < row['dog-automobile']:
        return 1
    else:
        return 0
```

```python
#Computing the number of correct predictions for ‘dog’:
dog_prediction = dog_distances.apply(is_dog_correct)
dog_prediction.head()
```

```python
dog_distances[4] #check
```

```python
# Accuracy of predicting dog in the test : 68%
print "{:.10f}%".format(float(dog_prediction.sum())/float(len(dog_prediction)))
```
